```python
# Question 9 


# turn this into a function and then add signals

# Signal params for 50 Hz
f1 = 50      # 50 Hz Natural freq
Fs1 = 44100  # Sampling freq 
t1 = 0.03      # seconds of sampling

T_1 = 1/Fs1                # sampling period
N_1 = Fs1*t1                # Total No.of points in the signal

omega_50 = 2*np.pi*f1    # angular frequency for sine waves 
t_vec_50 = np.arange(N_1)*T_1  # time vector for plotting
y_50 = np.sin(omega_50*t_vec_50)


# Signal params for 700 Hz
f2 = 700      # 50 Hz Natural freq
Fs2 = 44100  # Sampling freq 
t2 = 0.03      # seconds of sampling

T_2 = 1/Fs2                # sampling period
N_2 = Fs2*t2                # Total No.of points in the signal

omega_700 = 2*np.pi*f2    # angular frequency for sine waves 
t_vec_700 = np.arange(N_2)*T_2  # time vector for plotting
y_700 = np.sin(omega_700*t_vec_700)


# Plotting sum of two sine frequency graphs
plt.plot((t_vec_50 + t_vec_700),(y_50 + y_700))
plt.ylabel('Amplitude')
plt.xlabel('time [s]')
plt.show()
```
